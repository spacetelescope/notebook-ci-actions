```python
# Import basic libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

print("Basic libraries imported successfully!")
print(f"NumPy version: {np.__version__}")
print(f"Pandas version: {pd.__version__}")
```

```python
# Create simple test data
data = np.random.randn(100)
df = pd.DataFrame({
    'values': data,
    'squared': data**2,
    'index': range(100)
})

print(f"Created DataFrame with {len(df)} rows")
print("\nFirst 5 rows:")
print(df.head())
```

```python
# Create a simple plot
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.hist(df['values'], bins=20, alpha=0.7)
plt.title('Distribution of Random Values')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.scatter(df['values'], df['squared'], alpha=0.6)
plt.title('Values vs Squared Values')
plt.xlabel('Original Value')
plt.ylabel('Squared Value')

plt.tight_layout()
plt.show()

print("✅ Basic validation test completed successfully!")
```
